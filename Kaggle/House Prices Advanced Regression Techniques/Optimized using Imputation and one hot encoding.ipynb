```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import Imputer


# Read the train data
train = pd.read_csv('train.csv')

# Read the test data
test = pd.read_csv('test.csv')

# pull data into target (y) and predictors (X)
train_y = train.SalePrice

predictors_X = train.drop(['SalePrice'],axis=1)

numeric_predictors_X = predictors_X.select_dtypes(exclude=['object'])
categorical_predictors_X = predictors_X.select_dtypes(include=['object'])

# Create training predictors data
train_X = numeric_predictors_X

# Treat the test data in the same way as training data. In this case, pull same columns.
test_X = test[numeric_predictors_X.columns]

# Imputation : Fills the missing with some numbers

my_imputer = Imputer()
imputed_X_train = my_imputer.fit_transform(train_X)

imputed_X_test = my_imputer.transform(test_X)
```

```python
# Converting Array to dataframe

def ArraytoDataFrame(array_dataset,columns):
    df = pd.DataFrame()
    for i in range(0,len(columns)):
        df[columns[i]] = array_dataset[i]
    return df
```

```python
# One Hot Encoding

training_predictors = categorical_predictors_X
testing_predictors = test[categorical_predictors_X.columns]
one_hot_encoded_training_predictors = pd.get_dummies(training_predictors)

one_hot_encoded_test_predictors = pd.get_dummies(testing_predictors)
final_train, final_test = one_hot_encoded_training_predictors.align(one_hot_encoded_test_predictors,
                                                                    join='left', 
                                                                    axis=1)
```

```python
# Join numeric and categorical datasets

# Array to dataframe

X_train = ArraytoDataFrame(imputed_X_train,numeric_predictors_X.columns) + final_train
X_test = ArraytoDataFrame(imputed_X_test,numeric_predictors_X.columns) + final_test
```

```python
my_model = RandomForestRegressor()
my_model.fit(X_train, train_y)

predicted_prices = my_model.predict(X_test)
# We will look at the predicted prices to ensure we have something sensible.
print(predicted_prices)
```
